```python
def kmers(sequence, size = 6):
    return [sequence[x:x+size] for x in range(len(sequence) - size + 1)]
```

```python
def extractingExons(seq):
    pattern = "(?:atg)[a-z]+?(?=taa|tag|tga)"
    return ''.join(re.findall(pattern, seq))
```

```python
def dnaToProtein(seq):
    
    table = {'ata': 'i', 'atc': 'i', 'att': 'i', 'atg': 'm',
             'aca': 't', 'acc': 't', 'acg': 't', 'act': 't',
             'aac': 'n', 'aat': 'n', 'aaa': 'k', 'aag': 'k',
             'agc': 's', 'agt': 's', 'aga': 'r', 'agg': 'r',
             'cta': 'l', 'ctc': 'l', 'ctg': 'l', 'ctt': 'l',
             'cca': 'p', 'ccc': 'p', 'ccg': 'p', 'cct': 'p',
             'cac': 'h', 'cat': 'h', 'caa': 'q', 'cag': 'q',
             'cga': 'r', 'cgc': 'r', 'cgg': 'r', 'cgt': 'r',
             'gta': 'v', 'gtc': 'v', 'gtg': 'v', 'gtt': 'v',
             'gca': 'a', 'gcc': 'a', 'gcg': 'a', 'gct': 'a',
             'gac': 'd', 'gat': 'd', 'gaa': 'e', 'gag': 'e',
             'gga': 'g', 'ggc': 'g', 'ggg': 'g', 'ggt': 'g',
             'tca': 's', 'tcc': 's', 'tcg': 's', 'tct': 's',
             'ttc': 'f', 'ttt': 'f', 'tta': 'l', 'ttg': 'l',
             'tac': 'y', 'tat': 'y', 'taa': '*', 'tag': '*',
             'tgc': 'c', 'tgt': 'c', 'tga': '*', 'tgg': 'w'}
    
    protein = "".join([table[seq[i:i + 3]] for i in range(0, len(seq), 3)])
    return protein
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def get_metrics(y_test, y_predicted):
    accuracy = accuracy_score(y_test, y_predicted)
    precision = precision_score(y_test, y_predicted, average='weighted')
    recall = recall_score(y_test, y_predicted, average='weighted')
    f1 = f1_score(y_test, y_predicted, average='weighted')
    return accuracy, precision, recall, f1

accuracy, precision, recall, f1 = get_metrics(y_test, y_pred)
print("accuracy = %.3f \nprecision = %.3f \nrecall = %.3f \nf1 = %.3f" %
(accuracy, precision, recall, f1))
```
